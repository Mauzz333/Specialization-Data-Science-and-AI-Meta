```python
#Diccionario de todo los productos

productos = {
    'Producto_1': {'Tipo':'Pizza','Tamaño': 'Mediana', 'Ventas' : 20, 'Descripción': 'Champiñon' },
    'Producto_2': {'Tipo':'Pizza','Tamaño': 'Chica', 'Ventas' : 50, 'Descripción': 'Champiñon'  },
    'Producto_3': {'Tipo':'Pizza','Tamaño': 'Grande', 'Ventas' : 40, 'Descripción': 'Champiñon'},
    'Producto_4': {'Tipo':'Pizza','Tamaño': 'Mediana', 'Ventas' : 30, 'Descripción': 'Peperoni' },
    'Producto_5': {'Tipo':'Pizza','Tamaño': 'Chica', 'Ventas' : 10, 'Descripción': 'Peperoni'  },
    'Producto_6': {'Tipo':'Pizza','Tamaño': 'Grande', 'Ventas' : 70, 'Descripción': 'Peperoni'},
    'Producto_7': {'Tipo':'Pizza','Tamaño': 'Mediana', 'Ventas' : 30, 'Descripción': 'Mixta' },
    'Producto_8': {'Tipo':'Pizza','Tamaño': 'Chica', 'Ventas' : 10, 'Descripción': 'Mixta'  },
    'Producto_9': {'Tipo':'Pizza','Tamaño': 'Grande', 'Ventas' : 70, 'Descripción': 'Mixta'},
    'Producto_10': {'Tipo':'Bebida','Tamaño': 'Mediana', 'Ventas' : 100, 'Descripción': 'Coca-Cola'},
    'Producto_11': {'Tipo':'Bebida','Tamaño': 'Chica', 'Ventas' : 60, 'Descripción': 'Coca-Cola'},
    'Producto_12': {'Tipo':'Bebida','Tamaño': 'Grande', 'Ventas' : 110, 'Descripción': 'Coca-Cola'},
    'Producto_13': {'Tipo':'Bebida','Tamaño': 'Grande', 'Ventas' : 20, 'Descripción': 'Agua simple'}
}
```

```python
#Funcion para ver todas las ventas
def ver_registros(productos):
  count = 0
  for producto in productos:
    count += 1
    print(f'Producto_{count}:')
    for clave, valor in productos[producto].items():
      print(f'{clave}:{valor}')

ver_registros(productos)

```

```python
#Funcion crear producto
def crear_registro(productos):
  try:
    tipo = str(input('Ingrese el tipo de productos(Pizza-Bebida-Burger-Hotdog): '))
    if tipo == 'Burger' or tipo == 'Hotdog':
      tamaño = 'N/A'
    else:
      tamaño = str(input('Ingrese el tamaño (Grande-Mediana-Chica): '))
    ventas = int(input('Ingrese las ventas obtenidas: '))
    descripcion = str(input('Ingrese descripción: '))

    new_producto = {'Tipo': tipo,'Tamaño': tamaño , 'Ventas' : ventas, 'Descripción': descripcion}

    productos[f'Producto_{len(productos)+1}'] = new_producto
    print('¡ Se ha registrado con éxito !')
    return new_producto

  except:
    print('Error al ingresar los datos, porfavor intente de nuevo.')

crear_registro(productos)
```

```python
#Función consultar registro
def consultar_registro():
  def menu_query():
    while True:
      print('\nMenu de consulta')
      print('1. Consultar producto')
      print('2. Consultar promedio de productos')
      print('3. Consultar total de ventas')
      print('4. Consultar máximo de ventas del producto')
      print('5. Consultar ganancias')
      print('6. Volver al menú principal')
      opcion = input('Seleccione una opción: ')
      if opcion == '1':
        buscador = input('Ingrese el nombre del producto (ej. Producto_1): ').strip()
        finded = False

        for producto in productos:
          if buscador == producto:
            print(f'El producto {buscador} se ha encontrado en la base de datos.')
            print(f'Tipo: {productos[buscador]["Tipo"]}')
            finded = True
            break
      elif opcion == '2':
        promedio = lambda productos: sum(producto['Ventas'] for producto in productos.values()) / len(productos)
        print(f'El promedio de ventas es: {promedio(productos)}')
      elif opcion == '3':
        total_ventas = lambda productos: sum(producto['Ventas'] for producto in productos.values())
        print(f'El total de ventas es: {total_ventas(productos)}')
      elif opcion == '4':
        max_ventas = lambda productos: max(producto['Ventas'] for producto in productos.values())
        print(f'El máximo de ventas es: {max_ventas(productos)}')
      elif opcion == '5':
        ganancias = lambda productos: sum(producto['Ventas'] * 120 for producto in productos.values())  # Suponiendo un precio fijo de 10 por venta
        print(f'Las ganancias totales son: {ganancias(productos)}')
      elif opcion == '6':
        break
        print('Saliendo del Menú de consulta . . .')
      else:
        print('Opción no válida, por favor intente de nuevo.')
  menu_query()
consultar_registro()
```

```python
def actualizar_registro():
  def menu_update():
    while True:
      print('\nMenu de actualización de todos los productos')
      print('1. Actualizar Tipo')
      print('2. Actualizar Tamaño')
      print('3. Actualizar Ventas')
      print('4. Actualizar Descripción')
      print('5. Volver al menú principal')


      opcion = input('Seleccione una opción: ').strip()
      if opcion == '1':
        actualizar_tipo()
        continue
      elif opcion == '2':
        actualizar_tamaño()
        continue
      elif opcion == '3':
        actualizar_ventas()
        continue
      elif opcion == '4':
        actualizar_descripcion()
        continue
      elif opcion == '5':
        print('Saliendo del Menú de actualización . . .')
        break

      else:
        print('Opción no válida, por favor intente de nuevo')
      break


  menu_update()
actualizar_registro()
```

```python
#Funcion Actualizar el tipo de producto
def actualizar_tipo():
    try:
        buscador = input('Ingrese el nombre del producto (ej. Producto_1): ').strip()

        if buscador in productos:
            print(f'El producto {buscador} se ha encontrado.')
            print(f'Tipo: {productos[buscador]['Tipo']}')

            nuevo_tipo = input('Ingrese el nuevo tipo de producto: ')
            productos[buscador]['Tipo'] = nuevo_tipo

            print(f'¡Éxito! El tipo del {buscador} ha sido actualizado a: {nuevo_tipo}.')
            return productos[buscador]
        else:
            print(f'Lo siento, el producto "{buscador}" no existe en la base de datos.')

    except:
        return None

actualizar_tipo()
```

```python
#Función Actualizar tamaño del producto
def actualizar_tamaño():
  try:
    buscador = input('Ingrese el nombre del producto ( Ej. Producto_1): ').strip()
    if buscador in productos:
      print(f'El producto {buscador} se encuentra en la base de datos')
      print(f'Tipo: {productos[buscador]["Tipo"]}')

      new_tamaño = str(input('Ingrese el tamaño del produto(Ej. Grande, Mediana, Chica o N/A): ')).strip()
      productos[buscador]['Tamaño'] = new_tamaño

      print(f'¡Éxito! El tamaño del {buscador} ha sido actualizado a: {new_tamaño}.')
      return productos[buscador]
    else:
      print(f'Lo siento, el producto "{buscador}" no existe en la base de datos.')

  except:
     return None

actualizar_tamaño()
```

```python
#Función Actualizar Ventas del producto
def actualizar_ventas():
  try:
    buscador = input('Ingrese el nombre del producto ( Ej. Producto_1): ').strip()
    if buscador in productos:
      print(f'El producto {buscador} se encuentra en la base de datos')
      print(f'Tipo: {productos[buscador]["Tipo"]}')

      new_sale = input('Ingrese las ventas del produto: ').strip()
      productos[buscador]['Ventas'] = new_sale
      print(f'¡Éxito! Las ventas del {buscador} ha sido actualizado a: {new_sale}.')
      return productos[buscador]

    else:
      print(f'Lo siento, el producto "{buscador}" no existe en la base de datos.')

  except:
     return None

actualizar_ventas()
```

```python
#Función Actualizar Descripción del Producto
def actualizar_descripcion():
  try:
    buscador = input('Ingrese el nombre del producto (Ej. Producto_1): ').strip()
    if buscador in productos:
      print(f'El producto {buscador} se encuentra en la base de datos')
      print(f'Tipo: {productos[buscador]["Tipo"]}')

      new_description = str(input('Ingrese la nueva descripción del producto: ')).strip()
      productos[buscador]['Descripción'] = new_description
      print(f'Excelente, La descripción del {buscador} ha sido actualizada a: {new_description}')
      return productos[buscador]
    else:
      print(f'Lo sentimos, el {buscador}, no se encuentra en la base de datos')

  except:
    return None
actualizar_descripcion()
```

```python
def eliminar_registro():
  try:
    buscador = input('Ingrese el nombre del producto (Ej. Producto_1): ').strip()
    finded = False

    for producto in productos:
      if buscador == producto:
        print(f'El producto {buscador} se ha encontrado en la base de datos.')
        print(f'Tipo: {productos[buscador]["Tipo"]}')
        finded = True
        productos.pop(buscador)
        print(f'¡Éxito! El producto {buscador} ha sido eliminado de la base de datos.')
        break

    if not finded:
      print(f'Lo sentimos, el {buscador}, no se encuentra en la base de datos.')

  except:
    return None
eliminar_registro()

```
