Una **función integrada** es un bloque de código reutilizable

## **FUNCIÓN DEFINIDA POR EL USUARIO**

Una **función definida por el usuario** es creada de manera personalizada para realizar tareas específicas y reutilizables en todo el programa. Este tipo de función se crea utilizando la palabra clave ` def` seguida del nombre de la función y los parámetros, si los hay.

Una función definida por el usuario tiene la siguiente estructura:

```
def nombre_funcion(parametro1, parametro2, ...):
    # Código que realiza una tarea específica

    # Devolver un valor (opcional)
    return valor
```

### **EJEMPLO 1:**

Función que calcula el cuadrado de un número:

````python
# difinir función
#from IPython.core.formatters import Integer
num_cuad = input('Ingrese un número')

def cuadrado(numero):
    numero = float(numero)
    return numero **2

cuadrado(num_cuad)
````

### **EJEMPLO 2:**

Función para verificar si un número es par o impar:

````python
# difinir función
num_par = int(input("Ingrese un número: "))
def par(numero):
    if numero % 2 == 0:
        return 'el numero es par'
    else:
        return 'el numero no es par'

par(num_par)
````

### **EJEMPLO 3:**

Función para calcular el factorial de un número:

````python
# difinir función
num_fact = int(input('Enter a number'))

def factorial(numero):
    if numero == 0:
        return 1
    else:
        return numero * factorial(numero -1) # 4! = 4 * 3 * 2! --> 2 * 1
factorial(num_fact)
````

##Uso de Try y Except

***ValueError***: Esta excepción se lanza cuando una función recibe un argumento del tipo correcto, pero con un valor inapropiado. Por ejemplo, si intentas convertir la cadena "hola" a un número entero con int("hola").

***TypeError***: Ocurre cuando una operación o función se aplica a un objeto de un tipo inapropiado. Imagina que intentas sumar un número a una cadena directamente, como 5 + "texto".

***ZeroDivisionError***: Como su nombre lo indica, salta cuando intentas dividir un número por cero.

***FileNotFoundError***: Muy común cuando tu programa intenta acceder a un archivo que no existe en la ruta especificada.

***IndexError***: Se produce cuando intentas acceder a un índice que está fuera del rango de una secuencia, como una lista. Por ejemplo, si tienes una lista de 3 elementos e intentas acceder al elemento en el índice 5.

***KeyError***: Esta es frecuente cuando trabajas con diccionarios y tratas de acceder a una clave que no existe en el diccionario.

***AttributeError***: Se lanza cuando intentas acceder a un atributo o método que no existe en un objeto. Por ejemplo, si tienes un objeto mi_objeto y llamas a mi_objeto.metodo_inexistente().

***NameError***: Aparece cuando intentas usar una variable que no ha sido definida.

````python
try:
    # Este es el bloque donde ponemos el código que podría fallar.
    # Vamos a forzar un error, por ejemplo, dividiendo por cero
    # o intentando acceder a un índice fuera de rango.
    # Ejemplo 1: Dividir por cero (ZeroDivisionError)
    resultado = 10 / 0
    # Ejemplo 2: Intentar acceder a un índice que no existe (IndexError)
    # lista = [1, 2, 3]
    # print(lista[5])
    # Ejemplo 3: Intentar usar una variable no definida (NameError)
    # print(variable_no_existente)
except Exception as e:
    # Este bloque se ejecuta si ocurre CUALQUIER tipo de excepción
    # dentro del bloque 'try'. 'Exception' es la clase base para la mayoría
    # de los errores, así que la usamos para atraparlos todos.
    # 'as e' nos permite capturar el objeto de la excepción,
    # que contiene información sobre el error.
    print(f"¡Oops! Se ha producido un error inesperado: {e}")
    print(f"El tipo de error es: {type(e)}")
print("El programa ha continuado después del manejo de errores.")
````
